## Session 2:Data Wrangling, Cleaning

```python
df = pd.read_csv(pre_path + "2019_Yellow_Taxi_Trip_Data.csv")
```

```python
df.head()
```

```python
df.describe().round(2)
```

```python
df.columns
```

```python
df.columns.str.contains('id$')
```

```python
df['store_and_fwd_flag'].value_counts()
```

```python
drop_cols = df.columns[df.columns.str.contains('id$|store_and_fwd_flag')]
drop_cols
```

```python
df_clean = df.drop(columns=drop_cols)
```

```python
df_clean = df_clean.rename(columns={
    'tpep_pickup_datetime': 'pickup_dt',
    'tpep_dropoff_datetime': 'dropoff_dt'
})
```

## Change Dtypes

```python
df_clean.dtypes
```

```python
sel_cols = ['pickup_dt', 'dropoff_dt']
df_clean[sel_cols] = df_clean[sel_cols].apply(pd.to_datetime)
df_clean[sel_cols]
```

```python
df_clean['pickup_dt'].dt.year
```

## Feature Engineering

```python
df_clean.head(2)
```

```python
df_clean['elapsed_time'] = df_clean['dropoff_dt'] - df_clean['pickup_dt']

df_clean['avg_speed'] = df_clean['trip_distance'] / df_clean['elapsed_time'].dt.total_seconds()/60/60

df_clean.head(3)
```

## Barplot

```python
df_p = df
df_p['month'] = df_p['date'].dt.month
df_p = df_p.pivot_table(
    index='month',
    columns='year',
    values='travelers'
)
df_p.head(2)
```

```python
sns.heatmap(df_p)
```

```python
sns.heatmap(df_p, cmap='Blues', annot=True, fmt='.10g')
```

```python
ax = df_p.plot(kind='bar', ylabel='travelers')

ax.legend(title='', loc='center', bbox_to_anchor=(0.5, -0.3), ncol=3, frameon=False)
```

## Session 4: Pandas Freestyle

```python
df_travel = pd.read_csv(pre_path + "tsa_melted_holiday_travel.csv")
```

```python
df_super = pd.read_csv(pre_path + "supermarket_sales.csv")
df_super['Product count'] = 1
df_super.head(5)
```

```python
df_super.dtypes
```

```python
df_super['report_dt'] = df_super['Date'].map(pd.to_datetime)
df_travel['report_dt'] = df_travel['date'].map(pd.to_datetime)
```

```python
df_travel_c = df_travel.query('year == 2019').dropna(subset=['holiday'])
df_travel_c.head()
```

```python
df_super.groupby(['Product line', ''])
```

```python
df_super.head(2)
```

```python
df_merge = pd.merge(
    df_super,
    df_travel_c,
    on = 'report_dt'
)
df_merge
```

```python
df_super.groupby(['report_dt']).agg({'Product count': 'sum'}).plot()
```

```python
df_plot = df_merge.pivot_table(index='holiday', columns='Product line', values='Product count', aggfunc='sum')
```

```python
df_plot.plot(kind='barh')
```

```python
df_plot.T.plot.barh()
```
